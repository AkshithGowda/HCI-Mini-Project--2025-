```python
# Load data and imports first (see code block 1)

# Get all 'fft_..._a' column names
fft_cols_a = [f'fft_{i}_a' for i in range(750)]

# Group by label and calculate the mean for each FFT column
avg_fft_a = data.groupby('label')[fft_cols_a].mean()

# Create the line plot
plt.figure(figsize=(16, 8))
plt.plot(avg_fft_a.loc['POSITIVE'], label='POSITIVE', color='green', alpha=0.7)
plt.plot(avg_fft_a.loc['NEGATIVE'], label='NEGATIVE', color='red', alpha=0.7)
plt.plot(avg_fft_a.loc['NEUTRAL'], label='NEUTRAL', color='blue', alpha=0.7)
plt.title('Average FFT Signal (Sensor A) by Emotion', fontsize=16)
plt.xlabel('FFT Feature Index (0-749)', fontsize=12)
plt.ylabel('Average Power', fontsize=12)
plt.legend()
plt.tight_layout()
plt.savefig('average_fft_plot.png')
```

```python
label_mapping = {'NEGATIVE': 0, 'NEUTRAL': 1, 'POSITIVE': 2}
```

```python
def preprocess_inputs(df):
    df = df.copy()
    
    df['label'] = df['label'].replace(label_mapping)
    
    y = df['label'].copy()
    X = df.drop('label', axis=1).copy()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=123)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preprocess_inputs(data)

```

```python
X_train
```

```python
inputs = tf.keras.Input(shape=(X_train.shape[1],))

expand_dims = tf.expand_dims(inputs, axis=2)

gru = tf.keras.layers.GRU(256, return_sequences=True)(expand_dims)

flatten = tf.keras.layers.Flatten()(gru)

outputs = tf.keras.layers.Dense(3, activation='softmax')(flatten)


model = tf.keras.Model(inputs=inputs, outputs=outputs)
print(model.summary())
```
